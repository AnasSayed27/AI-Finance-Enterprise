```python
from sqlalchemy import create_engine, Column, DateTime, Float, String, BigInteger
from sqlalchemy.orm import sessionmaker, declarative_base
from dotenv import load_dotenv
import os

load_dotenv('.env')

engine= create_engine(os.getenv('DATABASE_URL'))

Base= declarative_base()

class FinancialData(Base):
    __tablename__ = 'macro_financial_lake'

    date= Column(DateTime, primary_key= True)
    open= Column(Float)
    high= Column(Float)
    low= Column(Float)
    close= Column(Float)
    volume= Column(BigInteger)
    us_fed_rate= Column(Float)
    india_gdp= Column(Float)

Base.metadata.create_all(engine)
print('Table schema created successfully.')
```

```python
final_df.columns= [c.lower().replace(' ', '_') for c in final_df.columns]

try:
    final_df.to_sql(
        name= 'macro_financial_lake',
        con= engine,
        if_exists= 'replace',
        index= True,
        index_label= 'Date'
    )
    print('Data lake population completed.')
except Exception as e:
    print(e)
```

```python
from statsmodels.tsa.stattools import adfuller
import numpy as np

df= pd.read_sql('Select "Date", close from macro_financial_lake order by "Date"', engine)

df['log_returns']= np.log(df['close']/df['close'].shift(1))
df.dropna(inplace= True)

def adfuller_test(series, name):
    print(f"\n--- ADF Test for {name} ---")
    result= adfuller(series)
    print(f'ADF Statistic: {result[0]:.4f}')
    print(f'p-value: {result[1]:.4e}')
    if result[1] <= 0.05:
        print('Result: Stationary (Reject H0)')
    else:
        print('Result: Non-Stationary (Fail to Reject H0)')

adfuller_test(df['close'], 'Raw nifty prices')
adfuller_test(df['log_returns'], 'NIFTY Log Returns')
```

```python
fig, axes = plt.subplots(2, 1)

sns.lineplot(data=df, x=df.index, y='close', ax=axes[0])
axes[0].set_title('NIFTY Closing Prices')

sns.lineplot(data=df, x=df.index, y='log_returns', ax=axes[1])
axes[1].set_title('NIFTY Log Returns')

plt.tight_layout()
plt.show()
```

```python
rf_annual= 0.07
rf_daily= (1 + rf_annual) ** (1/252) - 1

excess_returns= df['log_returns'] - rf_daily
sharpe_ratio = (excess_returns.mean() / excess_returns.std()) * np.sqrt(252)

downside_return= df[df['log_returns'] < 0]['log_returns']
sortino_ratio = (excess_returns.mean() / downside_return.std()) * np.sqrt(252)

var_95 = df['log_returns'].mean() - (1.65 * df['log_returns'].std())

df['cum_return']= (1 + df['log_returns'].cumprod())
df['rolling_peak']= df['cum_return'].cummax()
df['drawdown']= (df['cum_return'] - df['rolling_peak']) / df['rolling_peak']
max_drawdown = df['drawdown'].min()

print(f"--- Risk Report: NIFTY 50 ---")
print(f"Annualized Sharpe Ratio: {sharpe_ratio:.2f}")
print(f"Annualized Sortino Ratio: {sortino_ratio:.2f}")
print(f"Daily 95% Value at Risk (VaR): {var_95:.2%}")
print(f"Maximum Drawdown (Historical): {max_drawdown:.2%}")
```
